```python
plt.figure(figsize = (10, 8))
plt.imshow(output_imageL)
plt.show()
plt.figure(figsize = (10, 8))
plt.imshow(output_imageR)
plt.show()
```

```python
x0 = int((bboxL[0][0]+bboxL[0][2])/2)
y0 = int((bboxL[0][1]+bboxL[0][3])/2)

x1 = int((bboxR[0][0]+bboxR[0][2])/2)
y1 = int((bboxR[0][1]+bboxR[0][3])/2)
```

```python
cv2.circle(output_imageL,(x0,y0),10,(255,0,0),-1)
plt.figure(figsize=(15,12))
plt.imshow(output_imageL)
```
