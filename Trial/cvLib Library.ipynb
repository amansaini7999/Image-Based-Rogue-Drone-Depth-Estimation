```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
imgL = cv2.imread('imgL.jpg', 0)
imgR = cv2.imread('imgR.jpg', 0)
```

```python
cv2.imshow('Left Image', imgL)
cv2.imshow('Right Image', imgR)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=11)
disparity = stereo.compute(imgL, imgR)
```

```python
disparity
```

```python
cv2.imshow('Disparity', disparity)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(disparity,'gray')
plt.show()
```

```python
img0 = cv2.imread('frame0119.jpg')
img1 = cv2.imread('frame0120.jpg')
```

```python
imgL = cv2.imread('imgL.jpg')
imgR = cv2.imread('imgR.jpg')
```

```python
bbox0, label0, conf0 = cv.detect_common_objects(img0)
bbox1, label1, conf1 = cv.detect_common_objects(img1)
```

```python
bboxL, labelL, confL = cv.detect_common_objects(imgL)
bboxR, labelR, confR = cv.detect_common_objects(imgR)
```

```python
output_image0 = draw_bbox(img0, bbox0, label0, conf0)
output_image1 = draw_bbox(img1, bbox1, label1, conf1)
```

```python
output_imageL = draw_bbox(imgL, bboxL, labelL, confL)
output_imageR = draw_bbox(imgR, bboxR, labelR, confR)
```

```python
cv2.imshow('Left Image', output_image0)
cv2.imshow('Right Image', output_image1)
cv2.imshow('Left Image (Bright)', output_imageL)
cv2.imshow('Right Image (Bright)', output_imageR)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (10, 8))
plt.imshow(imgL)
plt.show()
plt.figure(figsize = (10, 8))
plt.imshow(imgR)
plt.show()
```
